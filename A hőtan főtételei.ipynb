**1.** Három liter kétatomos $300\,K$-es gázt izobar módon melegítünk (légköri nyomáson) $500\,K$-re.
Mekkora a térfogati munka és a közölt hőmennyiség?

----
$\def\doubleunderline#1{\underline{\underline{#1}}}$
$
\begin{align*}
V_1 &= 3\,l \implies 3\,dm^3 = 3 \cdot 10^{-3}\,m^3 \\
T_1 &= 300\,K \\
p_1 &= 10^5\,Pa = p_2\\
T_2 &= 500\,K \\
f &= 5
\end{align*}
$

---

$
\begin{align*}
Q &= \dfrac{f}{2} \cdot N \cdot k \cdot \Delta T \\
\Delta T &= T_2 - T_1 =  500\,K - 300\,K = \underline{200\,K} \\
p_1 \cdot V_1 &= N \cdot k \cdot T_1 \\
N &= \dfrac{p \cdot V_1}{k \cdot T_1} \\
Q &= \dfrac{f}{2} \cdot \dfrac{p \cdot V_1}{k \cdot T_1} \cdot k \cdot \Delta T \\
Q &= \dfrac{f}{2} \cdot \dfrac{p \cdot V_1}{T_1} \cdot \Delta T = \dfrac{5}{2} \cdot \dfrac{10^5\,Pa \cdot 3 \cdot 10^{-3}\,m^3}{300\,K} \cdot 200\,K = \doubleunderline{500\,J}
\end{align*}
$

$ 300\,J $ a közölt hőmennyiség.

---
$
\begin{align*}
W &= p \cdot \Delta V \\
\dfrac{p_1 \cdot V_1}{T_1} &= \dfrac{p_2 \cdot V_2}{T_2} \\
V_2 &= \dfrac{V_1 \cdot T_2}{T_1} = \dfrac{3 \cdot 10^{-3}\,m^3 \cdot 500\,K}{300\,K} = \underline{5 \cdot 10^{-3}\,m^3} \\
\Delta V &= V_2 - V_1 = 5 \cdot 10^{-3}\,m^3 - 3 \cdot 10^{-3}\,m^3 = \underline{2 \cdot 10^{-3}\,m^3} \\
W &= 10^5\,Pa \cdot 2 \cdot 10^{-3}\,m^3 = \doubleunderline{200\,J}
\end{align*}
$

$ 200\,J $ a végzett térfogati munka.

**2.** A $20\,mol$ anyagmennyiségű, $p_1 = 10^5\,Pa$ nyomású, kétatomos ideális gázt $T_1 = 300\,K$ hőmérsékletről - a p-V diagramon látható módon - kétféleképpen melegítjük fel $T_2 = 350\,K$ hőmérsékletre.

![5](hotan_5.png)

$a)$ Mekkora lesz a melegítés végén az egyes esetekben a gáz nyomása és térfogata?  
$b)$ Mennyi hőt vesz fel melegítéskor a gáz az egyik és a másik esetben?

----

$
\begin{align*}
p_1 &= 10^5\,Pa \\
T_1 &= 300\,K \\
T_2 &= 350\,K \\
n &= 20\,mol \\
f &= 5
\end{align*}
$

----

$a)$

$
\begin{align*}
p_1 \cdot V_1 &= n \cdot R \cdot T_1 \\
V_1 &= \dfrac{n \cdot R \cdot T_1}{p_1} = \dfrac{20\,mol \cdot 8.314\,\tfrac{J}{mol \cdot K} \cdot 300\,K}{10^5\,Pa} = \underline{0.49884\,m^3}
\end{align*}
$

----
#### I. eset - Izochor

$a)$
$
\begin{align*}
V_2 &= V_1 = \doubleunderline{0.49\,m^3} \\
\dfrac{p_1 \cdot V_1}{T_1} &= \dfrac{p_2 \cdot V_2}{T_2} \\
p_2 &= \dfrac{p_1 \cdot T_2}{T_1} = \dfrac{10^5\,Pa \cdot 350\,K}{300\,K} = \doubleunderline{116\,666.67\,Pa} \\
\end{align*}
$

$b)$
$
\begin{align*}
\Delta T &= T_2 - T_1 = 350\,K - 300\,K = \underline{50\,K} \\
Q &= \Delta E_b = \dfrac{f}{2} \cdot n \cdot R \cdot \Delta T = \dfrac{5}{2} \cdot 20\,mol \cdot 8.314\,\tfrac{J}{mol \cdot K} \cdot 50\,K = \doubleunderline{20\,785\,J}
\end{align*}
$

----
#### II. eset - Izobar

$a)$
$
\begin{align*}
p_2 &= p_1 = \doubleunderline{10^5\,Pa} \\
\dfrac{p_1 \cdot V_1}{T_1} &= \dfrac{p_2 \cdot V_2}{T_2} \\
V_2 &= \dfrac{V_1 \cdot T_2}{T_1} = \dfrac{0.49\,m^3 \cdot 350\,K}{300\,K} = \doubleunderline{0.5717\,m^3}
\end{align*}
$

$b)$
$
\begin{align*}
\Delta V &= V_2 - V_1 = 0.5717\,m^3 - 0.49884\,m^3 = \underline{0.072827\,m^3} \\
W &= -p \cdot \Delta V = -10^5\,Pa \cdot 0.072827\,m^3 = \underline{-7\,282.65\,J} \\
\Delta E_b &= Q + W \\
Q &= \Delta E_b - W = 20\,785\,J - \left(-7282.65\,J\right) = \doubleunderline{28\,067.64\,J}
\end{align*}
$
