### Hűtés
- Hűtés célja
    - Élelmiszerek és
    - bomlékony biológiai anyagok felhasználhatóságának megnövelése
- Természetes hűtés: jéggel
    - A télen képződött jeget (folyóból vagy tóból kivágva) jól szigetelt árnyékos tárolókba, jégvermekbe rakták, itt még a forró nyári napokon is megmaradt.
- Mesterséges hűtés:
    - 1856-tól az iparban (James Harrison gőzkompressziós hűtése)
    - 1867-ben J. B. Sutherland elkészíti az első hűtött vasúti kocsit
    - 1913-ban a DOMELRE kifejleszti az első háztartási hűtőgépet
    - 1928-tól 1974-ig használtunk freon (CFC) gázokat a hűtési ciklusban hőcsereközegként
        - 1974-ben fedeztük fel a freon gázok káros hatását az ózonrétegre
    - 1939-ben megjelenik az első fagyasztó (Electrolux)
- Termoelektromos hűtés
    - Továbbá említésre méltó a Peltier-hatást kihasználó elektromos hőcsere eszköz.
        - Kisebb eszközök, kempingberendezések és űreszközök hűtésére használják
        - Előnye, hogy kicsi és nem igényel bonyolult gázvezetékeket vagy mozgó alkatrészeket

**875.** Az $ 5 \cdot 10^{-2}\,m^3 $ térfogatú oxigénpalackot $ 1.5 \cdot 10^7\,Pa $ nyomásra töltik fel, $ 15\,^{\circ}C $ hőmérsékleten.

$ a) $ Mennyi a palackba töltött oxigén tömege?

----
$ \def\doubleunderline#1{\underline{\underline{#1}}} $
$
\begin{align*}
V &= 5 \cdot 10^{-2}\,m^3 \\
p &= 1.5 \cdot 10^7\,Pa \\
T &= 15\,^{\circ}C = 288\,K \\ \\
M_{O_2} &= 32\,\tfrac{g}{mol}
\end{align*}
$

----
$
\begin{align*}
pV &= nRT \\
pV &= \dfrac{m}{M}RT \\
m &= \dfrac{MpV}{RT} = \dfrac{32\,\frac{g}{mol} \cdot 1.5 \cdot 10^7\,Pa \cdot 5 \cdot 10^{-2}\,m^3}{8.314\,\frac{J}{mol \cdot K} \cdot 288\,K} = 10\,023.25\,g = \doubleunderline{10.02\,kg}
\end{align*}
$

**876.** A napjainkban elérhető legkisebb vákuum $ 10^{-11}\,Pa $, $ 273\,K $ hőmérsékleten. Hány molekula, vagy atom van ilyen nyomáson $ 1\,cm^3 $-nyi gázban?

---

$
\begin{align*}
p &= 10^{-11}\,Pa \\
T &= 273\,K \\
V &= 1\,cm^3 = 10^{-6}\,m^3
\end{align*}
$

---

$
\begin{align*}
pV &= NkT \\
N &= \dfrac{pV}{kT} = \dfrac{10^{-11}\,Pa \cdot 10^{-6}\,m^3}{1.38\cdot 10^{-23}\,\frac{J}{K} \cdot 273\,K} = \doubleunderline{2\,654\,db}
\end{align*}
$

**895.** $ 2\,dm^3 $ normál állapotú levegővel $ 333\,J $ hőmennyiséget közlünk úgy, hogy térfogata állandó marad.  
Mekkora lesz a nyomása és hőmérséklete?

----

$
\begin{align*}
V_0 &= 2\,dm^3 = 2\cdot 10^{-3}\,m^3 = V_1 \\
Q &= 333\,J \\
T_0 &= 273\,K \\
p_0 &= 10^5\,Pa \\
f &= 5
\end{align*}
$

----
$
\begin{align*}
pV &= NkT \\
N &= \dfrac{pV}{kT} = \dfrac{10^5\,Pa \cdot 2\cdot 10^{-3}\,m^3}{1.38\cdot 10^{-23}\,\frac{J}{K} \cdot 273\,K} = \underline{5.3 \cdot 10^{22}} \\
\\
Q &= \dfrac{f}{2} \cdot N \cdot k \cdot \Delta T \\
\Delta T &= \dfrac{2 \cdot Q}{f\cdot N \cdot k} = \dfrac{2 \cdot 333\,J}{5 \cdot 5.3 \cdot 10^{22} \cdot 1.38\cdot 10^{-23}\,\frac{J}{K}} = \underline{181.818\,K} \\
T_1 &= T_0 + \Delta T = 273\,K + 181.818\,K = \doubleunderline{454.818\,K} \\ \\
\dfrac{p_0 \cdot V_0}{T_0} &= \dfrac{p_1 \cdot V_1}{T_1} \\
p_1 &= \dfrac{p_0 \cdot T_1}{T_0} = \dfrac{10^5\,Pa \cdot 454.818\,K}{273\,K} = \doubleunderline{166\,600\,Pa}
\end{align*}
$

**G. 733.** Egy kútból vizet húzunk fel. A kút mélysége $ 10 $ méter, a veder tömege $ 2\,kg $, a lánc tömege $ 3\,kg $, és a veder űrtartalma $ 12\,liter $. Mekkora a vízhúzás mechanikai hatásfoka? Függ-e a hatásfok a kút mélységétől?

m_{lánc} = 3\,kg \\
V = 12\,l = 12\,dm^3 \implies m_{víz} = 12\,kg
$

$$ F_{max} = \left(m_{vödör} + m_{víz}\right) \cdot g + F_{lánc_{max}} = \left(2\,kg + 12\,kg\right) \cdot 10\,\frac{m}{s^2} + 30\,N = \underline{170\,N} $$  
És legalább:
$$ F_{min} = \left(m_{vödör} + m_{víz}\right) \cdot g + F_{lánc_{min}} = \left(2\,kg + 12\,kg\right) \cdot 10\,\frac{m}{s^2} + 0\,N = \underline{140\,N} $$

**G. 734.** Egy gépkocsi 3 órán át $ 80\,\frac{km}{h} $ átlagsebességgel haladt. Három különböző útszakasz mindegyikén egyenletes sebességgel mozgott. A városi csúcs forgalomban $ v_1=20\,\frac{km}{h} $ volt a sebessége, az országúton $ v_2=80\,\frac{km}{h} $, és végül az autópályán másfél órán keresztül $ v_3=120\,\frac{km}{h} $ sebességgel haladt. Mennyi időt töltött az autó a csúcsforgalomban, és mennyi volt az autó átlagsebessége a városi és az országúti szakaszokon együttesen?

----

$
t_ö = 3\,h \\
v_{átl} = 80\,\tfrac{km}{h} \\
\\
v_1 = 20\,\tfrac{km}{h} \\
t_1 = ? \\
v_2 = 80\,\tfrac{km}{h} \\
v_3 = 120\,\tfrac{km}{h} \\
t_3 = 1.5\,h \\
v_{átl 1,2} = ?
$

----

$
\begin{align*}
s_ö &= v_{átl} \cdot t_ö = 80\,\tfrac{km}{h} \cdot 3\,h = \underline{240\,km} \\
s_3 &= v_3 \cdot t_3 = 120\,\tfrac{km}{h} \cdot 1.5\,h = \underline{180\,km} \\
\end{align*}
$

---

$
\begin{align*}
v_{átl 1,2} &= \dfrac{s_1 + s_2}{t_1 + t_2} \\
s_1 + s_2 &= s_ö - s_3 \\
s_1 + s_2 &= 240\,km - 180\,km \\
s_1 + s_2 &= \underline{60\,km} \\ \\
t_1 + t_2 &= t_ö - t_3 \\
t_1 + t_2 &= 3\,h - 1.5\,h \\
t_1 + t_2 &= \underline{1.5\,h} \\ \\
v_{átl 1,2} &= \dfrac{60\,km}{1.5\,h} = \doubleunderline{40\,\tfrac{km}{h}} \\
\end{align*}
$

A városi és országúton összesítve $ 40\,\tfrac{km}{h} $ volt az autó átlagsebessége.

$
v_1 \cdot t_1 + v_2 \cdot t_2 = s_1 + s_2 \\
\begin{cases}
20\,\tfrac{km}{h} \cdot t_1 + 80\,\tfrac{km}{h} \cdot t_2 &= 60\,km \\
t_1 + t_2 &= 1.5\,h \\
\end{cases} \\
$

$
\begin{align*}
t_2 &= 1.5\,h - t_1 \\
20\,\tfrac{km}{h} \cdot t_1 + 80\,\tfrac{km}{h} \cdot \left(1.5\,h - t_1\right) &= 60\,km \\
20\,\tfrac{km}{h} \cdot t_1 + 120\,km - 80\,\tfrac{km}{h} \cdot t_1 &= 60\,km \\
60\,\tfrac{km}{h} \cdot t_1 &= 60\,km \\
t_1 &= \doubleunderline{1\,h}
\end{align*}
$

A városi csúcsforgalomban egy órát töltött.
