**795.** Szívünk minden egyes összehúzódáskor $ 70~cm^3 $ vért továbbít az érrendszerbe. A szívből kiáramló és a szívbe bejutó vér nyomásának különbsége átlagosan $ 16~kPa $. Az átlagos pulzusszám $ \dfrac{72}{min} $.

$ a) $ Hány watt a szív átlagos teljesítménye?  
$ b) $ Mennyi zsír elégetése fedezi a szív napi munkáját, ha az $ 1~kg $ zsír elégetésekor keletkező $ 40~MJ $ energiát a szervezet $ 30\% $-os hatásfokkal hasznosítja?
$\def\doubleunderline#1{\underline{\underline{#1}}}$

-----

$
V = 70~cm^3 = 7 \cdot 10^{-5}~m^3 \\
\Delta p = 16~kPa \\
f = 1.2~\dfrac{1}{s} \implies t = \dfrac{1}{f} = \dfrac{1}{1.2}~s \\
$

----

$a)$
$
\begin{align*}
P &= \dfrac{W}{t} \\
W &= F \cdot s = p \cdot A \cdot s = p \cdot V \\
P &= \dfrac{W}{t} = \dfrac{p \cdot V}{\frac{1}{f}} = \dfrac{16\,000~Pa \cdot 7 \cdot 10^{-5}~m^3}{\frac{1}{1.2}~s} \\
P &= \doubleunderline{1.344~W}
\end{align*}
$

$b)$
$
\begin{align*}
E &= 1~kg~zsír \implies 40~MJ = 4\cdot 10^7~J \\
\eta &= 30\% = 0.3 = \dfrac{E_h}{E} = \dfrac{E_h}{4\cdot 10^7} \implies E_h = \underline{1.2 \cdot 10^7~J} \\
t &= 24 \cdot 60^2~s = 86\,400~s \\
E_n &= P \cdot t = 1.344~W \cdot 86\,400~s = \underline{116\,121.6~J} \\
\dfrac{E_n}{E_h} &= \dfrac{116\,121.6~J}{1.2 \cdot 10^7~J} = 0.096~kg = \doubleunderline{9.6~g}
\end{align*}
$

**804.** Egy $ 0~^{\circ}C $ hőmérsékletű jégtömbbe $ 150~\tfrac{m}{s} $ sebességű $ 1.2~kg $ tömegű és $ 30~^{\circ}C $ hőmérsékletű vasgolyó csapódik be. A jég olvadáshője $ 3.4 \cdot 10^5~\tfrac{J}{kg} $, a vas fajlagos hőkapacitása (fajhője) $ 460~\tfrac{J}{kg \cdot K} $.  
Legfeljebb mennyi jég olvad meg?

----

$
T_j = 0~^{\circ}C = 273~K \\
v = 150~\frac{m}{s} \\
m_v = 1.2~kg \\
T_g = 30~^{\circ}C = 303~K \\
L_o = 3.4 \cdot 10^5~\frac{J}{kg} \\
c = 460~\frac{J}{kg \cdot K} \\
m_j = ~?
$

---

$
\begin{align*}
Q = L_o \cdot m_j &= c \cdot m_v \cdot \Delta T + \dfrac{1}{2} m_v \cdot v^2 \\
3.4 \cdot 10^5~\frac{J}{kg} \cdot m_j &= 460~\frac{J}{kg \cdot K} \cdot 1.2~kg \cdot 30~K + \dfrac{1}{2} 1.2~kg \cdot \left(150~\frac{m}{s}\right)^2 \\
m_j &= \doubleunderline{0.088~kg}
\end{align*}
$

**807.** $ 20~kg $ tömegű, $ 0~^{\circ}C $ hőmérsékletű jég és $ 15~kg $ $ 0~^{\circ}C $ hőmérsékletű víz keverékét mennyi $ 100~^{\circ}C $-os vízgőz hozzákeverésével lehet $ 60~^{\circ}C $ hőmérsékletre melegíteni, ha $ c_v = 4200~\tfrac{J}{kg \cdot K} $, $ L_o = 3.34 \cdot 10^5~\tfrac{J}{kg} $, $ L_f = 2.25 \cdot 10^6~\tfrac{J}{kg} $?

----

$
m_j = 20~kg \\
T_j = 273~K \\
m_v = 15~kg \\
T_v = 273~K \\
T_g = 373~K \\
\Delta T = 60~K \\
c_v = 4200~\frac{J}{kg \cdot K} \\
L_o = 3.34 \cdot 10^5~\frac{J}{kg} \\
L_f = 2.25 \cdot 10^6~\frac{J}{kg} \\
m_g = ~?
$

----

$
\begin{align*}
Q_j = L_o \cdot m_j &= 3.34 \cdot 10^5~\frac{J}{kg} \cdot 20~kg = \underline{6.68 \cdot 10^6~J} \\
Q_v = c \cdot \left(m_j + m_v\right) \cdot \Delta T &= 4200~\frac{J}{kg \cdot K} \cdot 35~kg \cdot 60~K = \underline{8.82 \cdot 10^6~J} \\
Q_g = L_f \cdot m_g &= 2.25 \cdot 10^6~\frac{J}{kg} \cdot m_g \\
Q* = c \cdot m_g \cdot \Delta T' &= 4200~\frac{J}{kg\cdot K} \cdot 40~K \cdot m_g \\ \\
Q_j + Q_v &= Q_g + Q* \\
6.68 \cdot 10^6~J + 8.82 \cdot 10^6~J &= m_g \left(2.25 \cdot 10^6~J + 4200~\frac{J}{kg\cdot K} \cdot 40~K\right) \\
m_g &= \doubleunderline{6.41~kg}
\end{align*}
$

**G. 737.** Hány egyenlő részre kell vágni a $ 100~\Omega $-os ellenálláshuzalt, hogy a részeket párhuzamosan kapcsolva az eredő ellenállásuk $ 1~\Omega $ legyen?

----

$
R_0 = 100~\Omega
$

$\begin{align*}
\dfrac{1}{R_e} &= n \cdot \dfrac{1}{R} = n \cdot \dfrac{1}{\frac{R_0}{n}} \\
\dfrac{1}{R_e} &= n \cdot \dfrac{1}{\frac{R_0}{n}} \\
\dfrac{1}{R_e} &= n \cdot \dfrac{n}{R_0} \\
\dfrac{1}{R_e} &= \dfrac{n^2}{R_0} \\
n &= \sqrt{\dfrac{R_0}{R_e}} \\
n &= \sqrt{\dfrac{100~\Omega}{1~\Omega}} = \doubleunderline{10}
\end{align*}$

$Tíz$ egyenlő részre kell vágni az ellenálláshuzalt.

Ell.
$
\dfrac{1}{R_e} = 10 \cdot \dfrac{1}{10} = \dfrac{1}{1} \implies 1~\Omega
$
