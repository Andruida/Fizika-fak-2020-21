**819.** Egy levegőbuborék térfogata a tengerszinten $ 4\,mm^3 $. A tengervíz sűrűsége $ 1030\,\frac{kg}{m^3} $, a tengerszinten a levegő nyomása $ 10^5\,\frac{N}{m^2} $.

$ a) $ Mekkora a nyomás a tengerben $ 15\,m $ mélységben?  
$ b) $ Mekkora a levegőbuborék térfogata ebben a mélységben, ha a hőmérséklet változatlan?

----
$ \def\doubleunderline#1{\underline{\underline{#1}}} $
$
\begin{align*}
p_0 &= 10^5\,\frac{N}{m^2} \\
\varrho &= 1030\,\frac{kg}{m^3} \\
h &= 15\,m \\
V_0 &= 4\,mm^3
\end{align*}
$

----
$ a) $  
$
\begin{align*}
p &= p_0 + h \varrho g = 10^5\,\frac{N}{m^2} + 15\,m \cdot 1030\,\frac{kg}{m^3} \cdot 10 \frac{m}{s^2} \\
p_1 &= 254\,500\,Pa = \doubleunderline{254.5\,kPa}
\end{align*}
$

-----

$ b) $  
$
\begin{align*}
p_0 \cdot V_0 &= p_1 \cdot V_1 \\
V_1 &= \dfrac{p_0 \cdot V_0}{p_1} = \dfrac{10^5\,Pa \cdot 4\,mm^3}{254\,500\,Pa} \\
V_1 &= \doubleunderline{1.57\,mm^3}
\end{align*}
$

**839.** Egy gázhőmérőben levő gáz térfogata $ 0\,^{\circ}C $-on $ 500\,cm^3 $.  
Mekkora a mért hőmérséklet, ha a gáz térfogata $ 520\,cm^3 $, és a nyomása mindkét esetben $ 10^5 Pa $.

----

$
\begin{align*}
p &= áll. \\
T_0 &= 0\,^{\circ}C = 273\,K \\
V_0 &= 500\,cm^3 \\
V_1 &= 520\,cm^3
\end{align*}
$

----

$
\begin{align*}
\dfrac{V_0}{T_0} &= \dfrac{V_1}{T_1} \\
T_1 &= \dfrac{V_1 \cdot T_0}{V_0} = \dfrac{520\,cm^3 \cdot 273\,K}{500\,cm^3} \\
T_1 &= 283.92\,K = \doubleunderline{10.92\,^{\circ}C}
\end{align*}
$

**840.** $ 300\,l $ $ 27\,^{\circ}C $ hőmérsékletű $ 10^5\,Pa $ nyomású gáz először állandó nyomáson $ 200\,l $-rel tágul, másodszor állandó térfogaton a hőmérséklete $ 123\,^{\circ}C $-ra emelkedik.

$ a) $ Mekkora a gáz hőmérséklete az állandó nyomáson végbement állapotváltozásai végén?  
$ b) $ Mekkora a gáz nyomása az állandó térfogaton végbement állapotváltozás végén?

----

$
\begin{align*}
V_0 &= 300\,l = 300\,dm^3 \\
T_0 &= 27\,^{\circ}C = 300\,K \\
p_0 &= 10^5\,Pa \\
\\
V_1 &= 500\,l = 500\,dm^3 \\
T_1 &= \,? \\
p_1 &= p_0 \\
\\
V_2 &= V_1 \\
T_2 &= 123\,^{\circ}C = 396\,K \\
p_2 &= \,?
\end{align*}
$

----

$ a) $  

$
\begin{align*}
\dfrac{V_0}{T_0} &= \dfrac{V_1}{T_1} \\
T_1 &= \dfrac{V_1 \cdot T_0}{V_0} = \dfrac{500\,dm^3 \cdot 300\,K}{300\,dm^3} \\
T_1 &= \doubleunderline{500\,K}
\end{align*}
$


$ b) $

$
\begin{align*}
\dfrac{p_1}{T_1} &= \dfrac{p_2}{T_2} \\
p_2 &= \dfrac{p_1 \cdot T_2}{T_1} = \dfrac{10^5\,Pa \cdot 396\,K}{300\,K} \\
p_2 &= \doubleunderline{13\,200\,Pa}
\end{align*}

**844.** Mekkora lesz a nyomás a $ 30\,^{\circ}C $ hőmérsékletű munkahelyen megtöltött $ 2 \cdot 10^3\,Pa $ nyomású neongázt tartalmazó kisülési csőben, ha a hőmérséklet $ -10\,^{\circ}C $-ra csökken?

----

$
\begin{align*}
V &= áll. \\
T_0 &= 30\,^{\circ}C = 303\,K \\
p_0 &= 2 \cdot 10^3\,Pa \\
T_1 &= -10\,^{\circ}C = 263\,K
\end{align*}
$

----

$
\begin{align*}
\dfrac{p_0}{T_0} &= \dfrac{p_1}{T_1} \\
p_1 &= \dfrac{p_0 \cdot T_1}{T_0} = \dfrac{2 \cdot 10^3\,Pa \cdot 263\,K}{303\,K} \\
p_1 &= 1\,735.97\,Pa = \doubleunderline{1.736\,kPa}
\end{align*}
$
